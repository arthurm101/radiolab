# 4.6 Nyquist Window

```python
divisor = 5
sample_freq = (62.5 / divisor) * u.MHz 
dt = (1 / sample_freq).to(u.ns)

N_samples = 16000

# sampling_time = (N_samples * dt).to(u.s)
times = np.arange(0,N_samples) * dt.to(u.us)

# freqs = (np.fft.fftfreq(N_samples, d=dt)).to(u.MHz)
scaling_factor = 1.0 / np.iinfo(np.int16).max


power_05 = np.abs(np.fft.fft(data[3][:128]))**2
power_10 = np.abs(np.fft.fft(data[1][:128]))**2
power_15 = np.abs(np.fft.fft(data[7][:128]))**2
power_20 = np.abs(np.fft.fft(data[2][:128]))**2

powers = [power_05,power_10,power_15,power_20]

import math

def ordinal(n):
    #stackoverflow.com/questions/9647202/
    return "%d%s" % (n,"tsnrhtdd"[(math.floor(n/10)%10!=1)*(n%10<4)*n%10::4])

def dft_f_range(dft_samples = 2**7):
    delta_nu = (1 / (dft_samples * dt.to(u.s))).to(u.MHz) # Freq per dft bin
    return delta_nu.value * (np.arange(0,dft_samples) - dft_samples/2)



MHz_values = np.arange(5,25,5)

fig, axes = plt.subplots(2,2, figsize = (15,7.5)) 

PLOT_SCALE = 1e10 / 100

for i , ax in enumerate(axes.reshape(-1)):
    
    # Plot points
    ax.plot(dft_f_range(),
            np.fft.fftshift(powers[i]) / PLOT_SCALE,
            ls = 'None',
            marker = 'o',
            color = 'green',
            markersize = 5)
    
    # Plot baseline
    ax.axhline(0, color = 'black', ls = '--', alpha = 0.3)
    
    # Plot connecting lines
    ax.vlines(dft_f_range(),
              ymin = 0,
              ymax = np.fft.fftshift(powers[i]) / PLOT_SCALE,
              color = 'green',
              linewidth = 0.5)
    
    # remove ticks
    ax.set_xticks([])
    ax.set_yticks([])
    
    ax.set_ylim(-1,11)
    ax.set_xlim(-7,7)
    
    ax.text(0.5, 0.9,
            '%d MHz - %s Window' %(MHz_values[i], ordinal(i + 1)),
            horizontalalignment='center',
            verticalalignment='center',
            transform=ax.transAxes)
    
        # set ticks for left side of plotgrid
    if (i % 2) == 0:
        ax.yaxis.set_major_locator(MultipleLocator(2))
        ax.yaxis.set_minor_locator(MultipleLocator(0.5)) 
        ax.tick_params(which='major', length = 15)
        ax.tick_params(which='minor', length = 5)
    
   
    
#     # set ticks for bottom side of plotgrid
    if i > 1:

        ax.set_xticks(np.arange(-3,4))
        ax.xaxis.set_major_locator(MultipleLocator(2))
        ax.xaxis.set_minor_locator(MultipleLocator(0.5))
        ax.tick_params(which='major', length = 15)
        ax.tick_params(which='minor', length = 5)

    if i == 0:
        ax.set_ylabel(r'Scaled Power $[k \cdot V^2 \cdot s^2]$', x = 0, y = 0)
    
    
        
    if i == 2:
        ax.set_xlabel(r'Frequency $[MHz]$', x = 1, y = 0)
    
    if i == 1:
        ax.set_title(r'Various signals sampled in successive Nyquist windows', x = 0)
    
plt.subplots_adjust(hspace=0, wspace = 0)  

fig.savefig('nyquist_windows_3_6.pdf', dpi = 'figure', format = 'pdf', transparent = True)
plt.show()
```

```python
#ssb_delay = ugradio.pico.capture_data('500mV', divisor=5, dual_mode=True)
#ssb_nodelay = ugradio.pico.capture_data('500mV', divisor=5, dual_mode=True)
#np.save('ssb_ndlower1', ssb_nodelay)
```

```python
delay_upper = np.load('ssb_d1.npy')
no_delay_upper = np.load('ssb_nd1.npy')

delay_lower = np.load('ssb_dlower1.npy')
no_delay_lower = np.load('ssb_ndlower1.npy')

no_delay_upper_port1 = no_delay_upper[:16000]
no_delay_upper_port2 = no_delay_upper[-16000:] #real

no_delay_lower_port1 = no_delay_lower[:16000]
no_delay_lower_port2 = no_delay_lower[-16000:] #real

def complex(data1, data2):
    
    comp = np.zeros(len(data1), dtype = np.complex_)
    
    for i in range (len(data1)):
        comp[i] = data2[i] + data1[i] * 1j
    return comp
```

```python
# UPPER NO DELAY PLOTTING

nodelay_complex = complex(no_delay_upper_port1, no_delay_upper_port2)
fft_nodelay = np.fft.fft(nodelay_complex)
power_nodelay = abs(fft_nodelay) ** 2
#plt.plot(np.fft.fftshift(power_nodelay))
```
